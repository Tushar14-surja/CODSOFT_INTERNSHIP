## Load data

### Subtask:
Load the data from the specified CSV file into a pandas DataFrame.


**Reasoning**:
Load the data into a pandas DataFrame and display its head and info to understand the data.



```
import pandas as pd

df = pd.read_csv('creditcard.csv')
display(df.head())
display(df.info())
```

## Preprocessing

### Subtask:
Preprocess the data by handling missing values, scaling features, and preparing it for model training.


**Reasoning**:
Identify columns with missing values and drop rows containing them, then separate features and target, scale the features, and drop the 'Time' column as it's not directly relevant for identifying fraudulent transactions based on their features.



```
df.dropna(inplace=True)

X = df.drop(['Class', 'Time'], axis=1)
y = df['Class']

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

display(X_scaled_df.head())
display(y.head())
```

## Handle class imbalance

### Subtask:
Address the class imbalance in the dataset using techniques like oversampling or undersampling.


**Reasoning**:
Address the class imbalance by oversampling the minority class using SMOTE and then display the value counts of the resampled target to confirm the balance.



```
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled_df, y)

display(y_resampled.value_counts())
```

## Split data

### Subtask:
Split the preprocessed data into training and testing sets.


**Reasoning**:
Split the resampled data into training and testing sets using train_test_split with stratification.



```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, stratify=y_resampled, random_state=42)

display(X_train.shape)
display(X_test.shape)
display(y_train.value_counts())
display(y_test.value_counts())
```

## Train model

### Subtask:
Train a classification model (e.g., Logistic Regression, Random Forest) on the training data.


**Reasoning**:
Import the Logistic Regression model, instantiate it, and train it using the training data.



```
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)
```

## Evaluate model

### Subtask:
Evaluate the trained model's performance using appropriate metrics such as precision, recall, and F1-score on the testing data.


**Reasoning**:
Calculate and print the precision, recall, and F1-score of the trained model on the test set.



```
from sklearn.metrics import precision_score, recall_score, f1_score

y_pred = model.predict(X_test)

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```

## Prediction

### Subtask:
Use the trained model to make predictions on new, unseen data.


**Reasoning**:
Use the trained model to make predictions on the test data and display the first few predictions.



```
y_pred = model.predict(X_test)
print("First few predictions:")
print(y_pred[:10])
```
